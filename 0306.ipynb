```python
# 색종이 2
di = [-1,1,0,0]
dj = [0,0,-1,1]

N = 102 #양쪽  패딩
arr = [[0]*(N) for _ in range(N)]
T = int(input())
#배열
for _ in range(T):
    si, sj = map(int, input().split())
    si += 1; sj += 1
    for i in range(si, si+10):
        for j in range(sj, sj+10):
            arr[i][j] = 1

ans = 0
#판별
for i in range(1, N):
    for j in range(1, N):
        if arr[i][j] == 1:
        #변두리 cnt
            l_cnt = 0
            for k in range(4):
                ni = i + di[k]
                nj = j + dj[k]
                if arr[ni][nj] == 1:
                    l_cnt += 1
            # 변 += 1
            if l_cnt == 3:
                ans += 1

            #모서리 += 2
            if l_cnt == 2:
                ans += 2

print(ans)
```

```python
#당근 분류 1
T = int(input())
for test_case in range(1, T+1):
    N = int(input())
    lst = list(map(int, input().split()))
    # 오름차순 정렬 후
    lst.sort()

    minV = 1000

    for i in range(N-2): #소박스
        for j in range(i+1, N-1): #중박스
            if lst[i] != lst[i+1] and [j]!=lst[j+1]: # 같은 크기가 다른 박스에 들어가는 경우 제외
                A = i + 1
                B = j - i
                C = N - 1 - j
                if A*B*C != 0 and A <= N//2 and B<=N//2 and C<=N//2: #빈 박스 X, 초과 X
                    if minV > max(A,B,C) - min(A,B,C):
                        minV = max(A, B, C) - min(A, B, C)


    if minV == 1000: #포장할 수 없는 경우
        minV = -1

    print(f"#{test_case} {minV}")
```

```python
# 당근 슬라이스
T = int(input())
for test_case in range(1, T+1):
    N = int(input())
    lst = list(map(int, input().split()))
    # 오름차순 정렬 후
    lst.sort()

    #v 정렬
    v = [0] * (30+1)
    for i in lst:
        v[i] += 1

    ans = 1000
    for i in range(31-2):
        for j in range(i+31-1):
            A = sum(v[1:i+1])
            B = sum(v[i+1:j+1])
            C = sum(v[j+1:31])
            # 0이 아니면서, 초과 X
            if A*B*C != 0 and A <= N//2 and B <= N//2 and C <= N//2:
                if ans > max(A,B,C) - min(A,B,C):
                    ans = max(A,B,C) - min(A,B,C)

    if ans == 1000:
        ans = -1

    print(f"#{test_case} {ans}")
```
