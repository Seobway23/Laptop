```python
url = "https://www.naver.com"
response = requests.get(url)
# 문제 생겼을 때, 종료
response.raise_for_status()

#문서값, 첫번째, lxml 파서블? html.parser
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.title)
print(soup.title.get_text())
```

```python
job_titles = []
for job in soup.select("job"):
    title = job.select_one(".title").text.strip()
    job_titles.append(title)


print(job_titles)
```

```python
import requests
from bs4 import BeautifulSoup

webpage = requests.get("https://vmit.tistory.com/") #크롤링할 웹페이지 url.

soup = BeautifulSoup(webpage.content, "html.parser") #웹페이지 내용을 html.parser를 이용해 변수 soup에 담는다.

 

#print(soup) #변수 soup 출력. #해당 부분은 웹 구성 확인할 때 이미 사용했으므로 삭제한다.

title_list = soup.find_all('span', class_ = 'title') #span 태그 안의 class가 title 인 것을 찾아, 변수 title_list에 담는다.

 

for span in title_list: #반복문 작성
    div = span.select('div > span') #div 태그 안에 포함된 span 태그 내용을 변수 div에 담는다.
    print('span:',span)
    for l in span: #span이 존재하는 동안 반복한다.
        print(l.get_text()) #태그 속의 텍스트만을 반환하기 위해 get_text를 적용. (태그 제외)
```
